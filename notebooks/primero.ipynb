```python
import pandas as pd

# Ruta
ruta = "/Users/quique/Desktop/mlops-trading-tfg/data/raw/AAPL.csv"

# 1. Leer saltando 2 filas (elimina encabezados)
df = pd.read_csv(ruta, skiprows=2, header=None)

# 2. Poner nombres manualmente
df.columns = ['Date', 'Open', 'High', 'Low', 'Close', 'Volume']

# 3. Convertir Date a datetime y poner como √çNDICE
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# Mostrar
print(df.head())
print(f"\n‚úÖ Datos listos para an√°lisis temporal")
print(f"Filas: {len(df)}")
print(f"Columnas: {df.columns.tolist()}")
print(f"Rango: {df.index[0].date()} a {df.index[-1].date()}")

# Ejemplo de ventajas del √≠ndice:
print(f"\nüìÖ Ejemplo - Filtro por a√±o 2020:")
print(df.loc['2020'].head())
```
