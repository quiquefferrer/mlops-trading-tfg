```python
import pandas as pd

path = "../data/raw/AAPL.csv"

# Cargar sin asumir nada
raw = pd.read_csv(path)

print("Columnas en el CSV:", raw.columns.tolist())
print(raw.head(3))

# Intentar detectar columna fecha típica
date_col_candidates = [c for c in raw.columns if c.lower() in ["date", "datetime", "timestamp"]]
if not date_col_candidates:
    # A veces yfinance guarda la fecha en la primera columna sin nombre tipo "Unnamed: 0"
    date_col_candidates = [raw.columns[0]]

date_col = date_col_candidates[0]
print("Usando como fecha:", date_col)

# Parsear fechas
raw[date_col] = pd.to_datetime(raw[date_col], errors="coerce")
raw = raw.dropna(subset=[date_col]).sort_values(date_col)

# Normalizar nombres
raw.columns = [c.lower().strip().replace(" ", "_") for c in raw.columns]
date_col = date_col.lower().strip().replace(" ", "_")

# Renombrar a 'date' (estándar)
raw = raw.rename(columns={date_col: "date"})

# Quedarnos con OHLCV estándar si existen
cols = ["date", "open", "high", "low", "close", "volume", "adj_close"]
df = raw[[c for c in cols if c in raw.columns]].copy()

df.head()
```
